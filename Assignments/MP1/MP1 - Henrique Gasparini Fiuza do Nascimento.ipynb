## Simple classification

### Creating and training a model with Keras

```python
from keras.models import Sequential
from keras.layers import Dense, Conv2D, Flatten, MaxPooling2D
from keras import regularizers
from keras import optimizers
```

```python
import contextlib, os
```

```python
def create_and_train_model(optimizer='sgd', l=0.0001):
    model = Sequential()
    model.add(Dense(3, input_shape=(IMAGE_SIZE ** 2,), activation='softmax', kernel_regularizer=regularizers.l2(l=l)))
    model.compile(
        optimizer=optimizer,
        loss='categorical_crossentropy',
        metrics=['accuracy'],
    )
    model.fit(X_train, np_utils.to_categorical(Y_train, num_classes=3), validation_split=0.1, epochs=100, batch_size=32)
    return model
```

## A More Difficult Classification Problem

```python
[X_train, Y_train] = generate_dataset_classification(300, 20, True)
```

```python
[X_large_train, Y_large_train] = generate_dataset_classification(3000, 20, True)
```

```python
def create_and_train_convolutional_model(optimizer='adam', large_data=False):
    model = Sequential()
    model.add(Conv2D(16, kernel_size=5, activation='relu', input_shape=(IMAGE_SIZE, IMAGE_SIZE,1)))
    model.add(MaxPooling2D(pool_size=(2,2)))
    model.add(Flatten())
    model.add(Dense(3, activation='softmax'))
    model.compile(
        optimizer=optimizer,
        loss='categorical_crossentropy',
        metrics=['accuracy'],
    )
    if large_data:
        X, Y = X_large_train, Y_large_train
    else:
        X, Y = X_train, Y_train
    model.fit(X.reshape(-1, IMAGE_SIZE, IMAGE_SIZE, 1), np_utils.to_categorical(Y, num_classes=3), validation_split=0.1, epochs=100, batch_size=32)
    return model
```

```python
model.evaluate(X_test.reshape((-1, IMAGE_SIZE, IMAGE_SIZE, 1)), Y_test)
```

### Comments

We observed that the Adam optimizer produced results siginificantly betten than SGD.

We show above the results for a training set with 300 examples, which gives a 71% accuracy.

With a larger training set with 3000 elements we could get to 92% accuracy. In fact, we probably can get as close to 100% accuracy as we want by using even larger training sets.

### Sort and normalize the output

#### Sort according to the angle with the X-axis

```python
def sort_points(Y):
    xs = Y[0::2]
    ys = Y[1::2]
    points = list(zip(xs, ys))
    points.sort(key=lambda p: (p[1]/p[0], p[0]))
    return [z for p in [list(p) for p in points] for z in p]
```

```python
Y_train = np.array([sort_points(Y) for Y in Y_train])
```

```python
Y_test = np.array([sort_points(Y) for Y in Y_test])
```

```python
normalizer = StandardScaler()
Y_train_normalized = normalizer.fit_transform(Y_train)
Y_test_normalized = normalizer.transform(Y_test)
```

### Study a more complex and more efficient model

```python
def create_and_train_complex_regression_model():
    model = Sequential()
    model.add(Conv2D(40, kernel_size=5, input_shape=(IMAGE_SIZE, IMAGE_SIZE, 1), activation='relu'))
    model.add(Dropout(0.3))
    model.add(Conv2D(20, kernel_size=4, activation='relu'))
    model.add(Dropout(0.3))
    model.add(Flatten())
    model.add(Dense(6))
    model.compile(
        optimizer='adam',
        loss='mean_squared_error'
    )
    model.fit(X_train.reshape((-1, IMAGE_SIZE, IMAGE_SIZE, 1)), Y_train_normalized, validation_split=0.1, epochs=100, batch_size=32)
    return model
```

## Image denoising

### Generating training and test data

```python
[X_train, Y_train] = generate_dataset_denoising(5000, free_location=True)
```

```python
[X_test, Y_test] = generate_test_set_denoising()
```

```python
X_train = X_train.reshape((-1, IMAGE_SIZE, IMAGE_SIZE, 1))
Y_train = Y_train.reshape((-1, IMAGE_SIZE, IMAGE_SIZE, 1))
X_test = X_test.reshape((-1, IMAGE_SIZE, IMAGE_SIZE, 1))
Y_test = Y_test.reshape((-1, IMAGE_SIZE, IMAGE_SIZE, 1))
```

```python
from keras.layers import BatchNormalization, UpSampling2D, Activation
from keras.layers.convolutional import ZeroPadding2D
```

### Solving image denoising as a regression problem

```python
def create_and_train_denoising_model_as_a_regression_problem():
    
    model = Sequential()

    model.add(Conv2D(60,kernel_size=(3,3),input_shape=(72,72,1,),padding='same'))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(Conv2D(40,kernel_size=(2,2),padding='same'))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(MaxPooling2D(pool_size=(2,2)))
    model.add(Conv2D(60,kernel_size=(3,3),padding='same'))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(Conv2D(40,kernel_size=(2,2),padding='same'))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(MaxPooling2D(pool_size=(2,2)))
    model.add(Conv2D(60,kernel_size=(3,3),padding='same'))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(Conv2D(20,kernel_size=(2,2),padding='same'))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(UpSampling2D(size=(2,2)))
    model.add(Conv2D(40,kernel_size=(3,3),padding='same'))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(Conv2D(20,kernel_size=(2,2),padding='same'))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(UpSampling2D(size=(2,2)))
    model.add(ZeroPadding2D(padding=(1, 1)))
    model.add(Conv2D(1,kernel_size=(3,3)))
    model.compile(loss='mean_squared_error',
    optimizer='adam')        
    
    model.fit(X_train, Y_train, validation_split=0.1, epochs=50, batch_size=32)
    return model
```

```python
model_1 = create_and_train_denoising_model_as_a_regression_problem()
```

```python
x = X_test[0]
```

```python
y = model_1.predict(x.reshape(1, IMAGE_SIZE, IMAGE_SIZE, 1))[0]
```

```python
plt.imshow(x.reshape(IMAGE_SIZE,IMAGE_SIZE), cmap='gray')
```

```python
plt.imshow(y.reshape(IMAGE_SIZE,IMAGE_SIZE), cmap='gray')
```

```python
model_1.evaluate(X_test.reshape(-1, IMAGE_SIZE, IMAGE_SIZE, 1), Y_test.reshape(-1, IMAGE_SIZE, IMAGE_SIZE, 1))
```

### Solving image denoising as a segmentation problem

```python
def create_and_train_denoising_model_as_a_segmentation_problem():
    
    model = Sequential()

    model.add(Conv2D(60,kernel_size=(3,3),input_shape=(72,72,1,),padding='same'))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(Conv2D(40,kernel_size=(2,2),padding='same'))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(MaxPooling2D(pool_size=(2,2)))
    model.add(Conv2D(60,kernel_size=(3,3),padding='same'))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(Conv2D(40,kernel_size=(2,2),padding='same'))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(MaxPooling2D(pool_size=(2,2)))
    model.add(Conv2D(60,kernel_size=(3,3),padding='same'))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(Conv2D(20,kernel_size=(2,2),padding='same'))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(UpSampling2D(size=(2,2)))
    model.add(Conv2D(40,kernel_size=(3,3),padding='same'))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(Conv2D(20,kernel_size=(2,2),padding='same'))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(UpSampling2D(size=(2,2)))
    model.add(ZeroPadding2D(padding=(1, 1)))
    model.add(Conv2D(1,kernel_size=(3,3)))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(loss='binary_crossentropy',
                  optimizer='adam',
                  metrics=['accuracy'])
    
    model.fit(X_train, Y_train, validation_split=0.1, epochs=50, batch_size=32)
    return model
```

```python
model_2 = create_and_train_denoising_model_as_a_segmentation_problem()
```

```python
x = X_test[0]
```

```python
y = model_2.predict(x.reshape(1, IMAGE_SIZE, IMAGE_SIZE, 1))[0]
```

```python
plt.imshow(x.reshape(IMAGE_SIZE,IMAGE_SIZE), cmap='gray')
```

```python
plt.imshow(y.reshape(IMAGE_SIZE,IMAGE_SIZE), cmap='gray')
```

```python
model_2.evaluate(X_test.reshape(-1, IMAGE_SIZE, IMAGE_SIZE, 1), Y_test.reshape(-1, IMAGE_SIZE, IMAGE_SIZE, 1))
```

### Comments

The second approach generates much better images. In fact, hourglass networks are more often used for segmentation problem. In our problem, the denoised image should have only black and white pixels, so we can handle the problem as a segmentation problem.

In practice, we get a very low value for the cross entropy loss and a 99.99% classification accuracy, using 5000 training examples.

The first approach (with regression) also produces interesting images, which are less noisy than the original ones. In fact, the gray regions that should be white share an uniform gray color. However, the image still is quite noisy close to the boundaries of the object.
